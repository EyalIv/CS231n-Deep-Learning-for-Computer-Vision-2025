## Inline Question 1:

We've only asked you to implement ReLU, but there are a number of different activation functions that one could use in neural networks, each with its pros and cons. In particular, an issue commonly seen with activation functions is getting zero (or close to zero) gradient flow during backpropagation. Which of the following activation functions have this problem? If you consider these functions in the one dimensional case, what types of input would lead to this behaviour?
1. Sigmoid
2. ReLU
3. Leaky ReLU

$\color{blue}{\textit Your Answer:}$

The activation functions that suffer from zero or near-zero gradient flow are Sigmoid and ReLU.

Sigmoid: This function suffers from the vanishing gradient problem. For inputs with a large absolute value (either very positive or very negative, e.g., x→∞ or x→−∞), the function saturates, and its derivative approaches zero. This effectively stops the flow of gradients through the neuron, halting learning.

ReLU: This function suffers from the "dying ReLU" problem. For any negative input (x≤0), the function's output is zero, and its derivative is exactly zero. This means no gradient will flow backward through the neuron for that data point, potentially causing the neuron to become permanently inactive and stop learning.

Leaky ReLU, by design, mitigates this issue by having a small, non-zero gradient for negative inputs, thus preventing the gradient flow from completely stopping.

````python
best_model = None



#################################################################################
# TODO: Tune hyperparameters using the validation set. Store your best trained  #
# model in best_model.                                                          #
#                                                                               #
# To help debug your network, it may help to use visualizations similar to the  #
# ones we used above; these visualizations will have significant qualitative    #
# differences from the ones we saw above for the poorly tuned network.          #
#                                                                               #
# Tweaking hyperparameters by hand can be fun, but you might find it useful to  #
# write code to sweep through possible combinations of hyperparameters          #
# automatically like we did on thexs previous exercises.                          #
#################################################################################
best_val_acc = 0
results = {}

# Define hyperparameter search space
hidden_sizes = [100, 200, 300]
learning_rates = [1e-4, 5e-4, 1e-3, 2e-3]
regularization_strengths = [0.0, 0.01, 0.1, 0.5]

print("Starting hyperparameter search...")
print("This may take a while...")

# Grid search over hyperparameters
for hidden_size in hidden_sizes:
    for learning_rate in learning_rates:
        for reg in regularization_strengths:
            print(f'\nTesting: hidden_size={hidden_size}, lr={learning_rate}, reg={reg}')

            # Create model
            input_size = 32 * 32 * 3
            num_classes = 10
            model = TwoLayerNet(input_size, hidden_size, num_classes)
            model.reg = reg

            # Create solver
            solver = Solver(model, data,
                          update_rule='sgd',
                          optim_config={'learning_rate': learning_rate},
                          lr_decay=0.95,
                          num_epochs=10,
                          batch_size=100,
                          print_every=1000,
                          verbose=False)

            # Train the model
            solver.train()

            # Get validation accuracy
            val_acc = solver.val_acc_history[-1]
            train_acc = solver.train_acc_history[-1]

            print(f'Validation accuracy: {val_acc:.4f}, Training accuracy: {train_acc:.4f}')

            # Store results
            results[(hidden_size, learning_rate, reg)] = {
                'val_acc': val_acc,
                'train_acc': train_acc,
                'model': model,
                'solver': solver
            }

            # Keep track of best model
            if val_acc > best_val_acc:
                best_val_acc = val_acc
                best_model = model
                print(f'*** New best validation accuracy: {best_val_acc:.4f} ***')

print(f'\nHyperparameter search complete!')
print(f'Best validation accuracy achieved: {best_val_acc:.4f}')

# Find and display the best hyperparameters
best_params = None
for params, result in results.items():
    if result['val_acc'] == best_val_acc:
        best_params = params
        break

if best_params:
    hidden_size, learning_rate, reg = best_params
    print(f'Best hyperparameters:')
    print(f'  Hidden size: {hidden_size}')
    print(f'  Learning rate: {learning_rate}')
    print(f'  Regularization: {reg}')

################################################################################
#                              END OF YOUR CODE                                #
################################################################################
````

1. Train on a larger dataset.

$\color{blue}{\textit Your Answer:}$

$\color{blue}{\textit Your Explanation:}$

The gap between training and testing accuracy indicates overfitting, where the model memorizes the training data but fails to generalize to new data. Two methods can help reduce this gap:

Train on a larger dataset - More training examples expose the model to greater diversity, helping it learn generalizable patterns rather than memorizing specific training instances.

Increase the regularization strength - Regularization (like L2 penalty) penalizes large weights, forcing the model to learn simpler, more generalizable patterns and preventing it from fitting noise in the training data.

Adding more hidden units would actually increase model capacity and likely worsen overfitting, making the gap larger rather than smaller.
